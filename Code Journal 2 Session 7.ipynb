```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create an array 'x' ranging from [0,1] in 100 steps
x = np.linspace(0,1,100)
```

```python
#define a function that returns exp(x)
def exp_function(x):
    return np.exp(x)
```

```python
#use the function to set y=exp(x)
y = exp_function(x)
```

```python
#P lot x vs. y and label the x-axis as “Time [milliseconds]” and the y-axis as “Awesomeness”
plt.plot(x,y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.title("Exponential Growth of Awesomeness Over Time")
plt.savefig("awesomeness_plot.pdf")
plt.show()
```
