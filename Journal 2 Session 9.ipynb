```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#pull 1000 random numbers distributed uniformly between [0,1]
random_numbers = np.random.uniform(low=0.0, high=1.0, size=1000)
```

```python
#Histogram the random numbers into 100 bins
plt.figure(figsize=(8,6))
plt.hist(random_numbers, bins=100, color='skyblue', edgecolor='black')
plt.xlabel('Random Number Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Uniform Random Numbers')
plt.grid(True, linestyle='--', alpha=0.5)
```

```python
#save figure
plt.savefig('uniform_random_histogram.pdf')
plt.show()
```
