```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create an array 'x' ranging from [0,1] in 100 steps
x = np.linspace(0,1,100)
```

```python
#define a function for sin(x) and cos(x)
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

```python
y_sin = sin_function(x)
y_cos = cos_function(x)
```

```python
#multipanel plot (1 row, 2 columns), plotting sin(x) vs. x in the left panel and cos(x) vs. x in the right panel
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

#Left panel: sin(x)
ax1.plot(x, y_sin)
ax1.set_title("sin(x)")
ax1.set_xlabel("x")
ax1.set_ylabel("sin(x)")

#Right panel: cos(x)
ax2.plot(x, y_cos)
ax2.set_title("cos(x)")
ax2.set_xlabel("x")
ax2.set_ylabel("cos(x)")

#Adjust layout so labels don't overlap
plt.tight_layout()

#save figure as a PDF
plt.savefig("sin_cos_plot.pdf")

#show the figure
plt.show()
```
