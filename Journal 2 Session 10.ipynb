```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate 1000 random numbers from normal distribution
random_numbers = np.random.normal(loc=0.0, scale=1.0, size=1000)
```

```python
#create historgram with 1000 bins
plt.figure(figsize=(8,6))
plt.hist(random_numbers, bins=100, color='darkred', edgecolor='black')
plt.xlabel('Random Number Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Normally Distributed Random Numbers')
plt.grid(True, linestyle='--', alpha=0.5)
```

```python
#save figure
plt.savefig('normal_random_histogram.pdf')
plt.show()
```
